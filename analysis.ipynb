Find minimum maximum and average base pay

```python
df['BasePay'] = pd.to_numeric(df['BasePay'].replace('Not Provided', np.nan))
```

Replace all 'Not Provided' to NaN

```python
df = df.replace(['Not Provided', 'Not provided'], np.nan)
```

```python
df
```

Drop rows with 5 or more null values

```python
nan_count = df.isna().sum(axis=1)
df = df[nan_count <= 5 ]
```

```python
df
```

Find the employee with maximum basepay

```python
df[df['BasePay'].max()==df['BasePay']]
```

Find average employee basepay per year

```python
df.groupby('Year')['BasePay'].mean()
```

Find the average basepay of employees per jobtitle

```python
df_sorted = df.groupby('JobTitle')['BasePay'].mean()
```

```python
df_sorted = df_sorted.sort_values(ascending=False).head(15)
df_sorted
```
